### 2-1 Renaming columns 

```python
df = pd.read_csv('manager_salary_survey.csv')
class ColumnRenamerHandler:
    def __init__(self, old_column_name, new_column_name, next_handler=None):
        self.old_column_name = old_column_name
        self.new_column_name = new_column_name
        self.next_handler = next_handler

    def set_next_handler(self, handler):
        self.next_handler = handler

    def handle_request(self, df):
        if self.old_column_name in df.columns:
            df.rename(columns={self.old_column_name: self.new_column_name}, inplace=True)
        if self.next_handler:
            df = self.next_handler.handle_request(df)
        return df

class ChainColumnRenamerHandler:
    def __init__(self):
        self.head_handler = None
        self.current_handler = None

    def add_handler(self, handler):
        if self.head_handler is None:
            self.head_handler = handler
            self.current_handler = handler
        else:
            self.current_handler.set_next_handler(handler)
            self.current_handler = handler

    def handle_request(self, df):
        if self.head_handler:
            return self.head_handler.handle_request(df)
        return df

class ColumnRenamerClient:
    def __init__(self, rename):
        self.handler_chain = ChainColumnRenamerHandler()
        self.build_handler_chain(rename)

    def build_handler_chain(self, rename_specs):
        for old, new in rename_specs:
            handler = ColumnRenamerHandler(old, new)
            self.handler_chain.add_handler(handler)

    def rename_columns(self, df):
        return self.handler_chain.handle_request(df)
    
    
    
    
    
rename = [
       ("How old are you?", "age"),
    ("What industry do you work in?", "industry"),
    ("Job title", "job_title"),
    ("If your job title needs additional context, please clarify here:", "job_title2"),
    ("What is your annual salary? (You'll indicate the currency in a later question. If you are part-time or hourly, please enter an annualized equivalent -- what you would earn if you worked the job 40 hours a week, 52 weeks a year.)", "annual_salary"),
    ("How much additional monetary compensation do you get, if any (for example, bonuses or overtime in an average year)? Please only include monetary compensation here, not the value of benefits.", "additional_salary"),
    ("Please indicate the currency", "currency"),
    ('If "Other," please indicate the currency here: ', "currency2"),
    ("If your income needs additional context, please provide it here:", "income_details"),
    ("What country do you work in?", "country"),
    ("If you're in the U.S., what state do you work in?", "state"),
    ("What city do you work in?", "city"),
    ("How many years of professional work experience do you have overall?", "overall_experience"),
    ("How many years of professional work experience do you have in your field?", "field_experience"),
    ("What is your highest level of education completed?", "education_level"),
    ("What is your gender?", "gender"),
    ("What is your race? (Choose all that apply.)", "race")
]

column_renamer_client = ColumnRenamerClient(rename)
df = column_renamer_client.rename_columns(df)
print(df.head(5))
```

The above code function is to rename our columns to make it easier for us while cleaning them.

Functions of each class:

ColumnRenamerHandler class is a part of the Chain of Responsibility design pattern, it is funcion is to rename a columns in a dataset.
ChainColumnRenamerHandler class function is to manage a chain of handlers for renaming the column.
ColumnRenamerClient class function is to provide an interface to develop and execute the chain of handlers for renaming the columns.

Timestamp:

```python

class Handler0:
    def __init__(self, column, next_handler=None):
        self.column = column
        self.next_handler = next_handler

    def set_next_handler(self, handler):
        self.next_handler = handler

    def handle_request(self, df):
        df[self.column] = pd.to_datetime(df[self.column], errors='coerce')
        erroneous_indices = df[df[self.column].isna()].index
        most_occuring_value = df[self.column].mode().iloc[0]
        df.loc[erroneous_indices, self.column] = most_occuring_value
        
        if self.next_handler:
            df = self.next_handler.handle_request(df)
        return df

class ChainCleaningHandler:
    def __init__(self):
        self.head_handler = None
        self.current_handler = None

    def add_handler(self, handler):
        if self.head_handler is None:
            self.head_handler = handler
            self.current_handler = handler
        else:
            self.current_handler.set_next_handler(handler)
            self.current_handler = handler

    def handle_request(self, df):
        if self.head_handler:
            return self.head_handler.handle_request(df.copy())
        return df

class CleaningClient:
    def __init__(self, columns_to_clean):
        self.handler_chain = ChainCleaningHandler()
        self.build_handler_chain(columns_to_clean)

    def build_handler_chain(self, columns_to_clean):
        for column in columns_to_clean:
            handler = Handler0(column)
            self.handler_chain.add_handler(handler)

    def clean_timestamps(self, df):
        return self.handler_chain.handle_request(df)




columns_to_clean = ['Timestamp']
timestamp_cleaning = CleaningClient(columns_to_clean)
df = timestamp_cleaning.clean_timestamps(df)

print(df)
print(df['Timestamp'].dtypes)
```

The only cleaning I have done for this variable is changing the type of the column timetstamp and replacing any erroneous or missing values by the most occuring value in our column. 

Functions of each class: 

Handler0 class function is to clean timestamp columns.
ChainCleaningHandler class function is to manage a chain of handlers to clean timestamp columns.
CleaningClient class function is to provide an interface to develop and execute the chain of handlers to clean the column.

In the variable age I have replaced the ranges of ages by integers by calculation the mean of each of the ranges. I tried to handle missing and erroneous values by replacing them with the most occuring age value, and I have converted the column type to integer.

```python
class Handler1:
    def __init__(self, column_name, column_mapping, default_value=None, next_handler=None):
        self.column_name = column_name
        self.column_mapping = column_mapping
        self.default_value = default_value
        self.next_handler = next_handler

    def set_next_handler(self, handler):
        self.next_handler = handler

    def handle_request(self, df):
        df[self.column_name] = df[self.column_name].fillna(self.default_value)
        df[self.column_name] = df[self.column_name].map(self.column_mapping).fillna(self.default_value)
        df[self.column_name] = pd.to_numeric(df[self.column_name], errors='coerce').astype('Int64')

        if self.next_handler:
            df = self.next_handler.handle_request(df)

        return df



class AgeProcessingClient(Handler1):
    def __init__(self, age_column, default_value=29):
        column_mapping = {
            'under 18': 17,
            '18-24': 21,
            '25-34': 29,
            '35-44': 39,
            '45-55': 50,
            '55-64': 60,
            '65 or over': 65
        }
        super().__init__(age_column, column_mapping, default_value=default_value)




age_column = 'age'
default_age_value = 29
age_processing_client = AgeProcessingClient(age_column, default_value=default_age_value)
age_processing_client.handle_request(df)
print(df)
print(df[age_column].dtypes)

```

Functions of each class:

Handler1 is a class handler in the chain of responsibility, and this class function is to process the column.  
AgeProcessingClient is a class that inherits from the first class and it processes age column more specifically.

```python
class Handler2:
    def __init__(self, next_handler=None):
        self.next_handler = next_handler

    def add_handler(self, handler):
        self.next_handler = handler

    def handle_request(self, df):
        self.write_unique_values(df)
        self.replace_na_values(df)
        if self.next_handler:
            self.next_handler.handle_request(df)
        return df

    def write_unique_values(self, df):
        pass

    def replace_na_values(self, df):
        pass


class UniqueValuesWriterHandler(Handler2):
    def __init__(self, column_name, file_prefix='unique_values', next_handler=None):
        super().__init__(next_handler)
        self.column_name = column_name
        self.file_prefix = file_prefix

    def write_unique_values(self, df):
        unique_values_file = f'{self.file_prefix}_{self.column_name}.txt'
        unique_values = df[self.column_name].unique()

        with open(unique_values_file, 'w', encoding='utf-8') as file:
            for val in unique_values:
                file.write(str(val) + '\n')
            print(f"Unique values in the '{self.column_name}' column written to '{unique_values_file}'")


class ReplaceNAHandler(Handler2):
    def __init__(self, column_name, replacement_value='default_replacement', next_handler=None):
        super().__init__(next_handler)
        self.column_name = column_name
        self.replacement_value = replacement_value

    def replace_na_values(self, df):
        df[self.column_name].fillna(self.replacement_value, inplace=True)





column_name = 'industry'
replacement_value = 'tech and computing'

handler_chain = UniqueValuesWriterHandler(column_name, file_prefix='unique_values')
handler_chain.add_handler(ReplaceNAHandler(column_name, replacement_value))
handler_chain.handle_request(df)

```

```python
class Handler3:
    def __init__(self, next_handler=None):
        self.next = next_handler

    def handle_request(self, df):
        if self.next is not None:
            df = self.next.handle_request(df)
        return df




class ColumnToFloatConverter(Handler3):
    def __init__(self, column_name, next_handler=None):
        super().__init__(next_handler)
        self.column_name = column_name

    def handle_request(self, df):
        df[self.column_name] = pd.to_numeric(df[self.column_name], errors='coerce').astype('float')
        return super().handle_request(df)



class NAAndFloatConversionHandler:
    def __init__(self, column_name, replacement_value='default_replacement'):
        self.handler_chain = ReplaceNAHandler(column_name, replacement_value)
        self.handler_chain.next = ColumnToFloatConverter(column_name)
    def process(self, df):
        return self.handler_chain.handle_request(df)




column_name_to_process = 'annual_salary'
replacement_value = 125498.51
handler_chain = NAAndFloatConversionHandler(column_name_to_process, replacement_value)

df = handler_chain.process(df)
print(df[column_name_to_process])
```

```python
class Handler4:
    def __init__(self, next_handler=None):
        self.next_handler = next_handler

    def set_next_handler(self, handler):
        self.next_handler = handler

    def handle_request(self, df):
        if self.next_handler:
            return self.next_handler.handle_request(df)
        return df


class UniqueValuesWriterHandler(Handler4):
    def __init__(self, column_name, file_prefix='unique_values', next_handler=None):
        super().__init__(next_handler)
        self.column_name = column_name
        self.file_prefix = file_prefix

    def handle_request(self, df):
        unique_values_file = f'{self.file_prefix}_{self.column_name}.txt'
        unique_values = df[self.column_name].unique()

        with open(unique_values_file, 'w', encoding='utf-8') as file:
            for val in unique_values:
                file.write(str(val) + '\n')
            print(f"Unique values in the '{self.column_name}' column written to '{unique_values_file}'")

        return super().handle_request(df)


class ReplaceNAHandler(Handler4):
    def __init__(self, column_name, replacement_value, next_handler=None):
        super().__init__(next_handler)
        self.column_name = column_name
        self.replacement_value = replacement_value

    def handle_request(self, df):
        df[self.column_name].fillna(self.replacement_value, inplace=True)
        return super().handle_request(df)


column_name = 'currency'
replace_na_handler = ReplaceNAHandler(column_name, replacement_value='USD')
unique_values_writer_handler = UniqueValuesWriterHandler(column_name)
replace_na_handler.set_next_handler(unique_values_writer_handler)
df = replace_na_handler.handle_request(df)
print(df[column_name])
```

I managed the erroneous and missing values by replacing them with the US dollar since it is the most common one. I added a code that will create a text file with all the unique values of currencies for further analysis. But I did not change the type of the column since it is already string.

Functions of each class:

Handler4 is a class handler in the chain of responsibility.
UniqueValuesWriterHandler is a class that inherits from Handler4 its function is to write unique values of the column to a text file.
ReplaceNAHandler is a class that inherits from Handler4 its function is to replace all na values in the column.

Save data in a clean csv file:

```python
class SaveCleanedDataHandler:
    def __init__(self, output_file, successor=None):
        self.output_file = output_file
        self.successor = successor

    def handle_request(self, cleaned_dataframe):
        self.save_to_file(cleaned_dataframe)
        if self.successor:
            self.successor.handle_request(cleaned_dataframe)

    def save_to_file(self, cleaned_dataframe):
        cleaned_dataframe.to_csv(self.output_file, index=False)
        print(f"Cleaned data saved to '{self.output_file}'.")


class DataFrameCleaner:
    def __init__(self, handlers=None):
        self.handlers = handlers
    def clean_data(self, dataframe):
        cleaned_dataframe = dataframe.copy()
        self.handle_request(cleaned_dataframe)
    def handle_request(self, cleaned_dataframe):
        if self.handlers:
            for handler in self.handlers:
                handler.handle_request(cleaned_dataframe)



cleaned = DataFrameCleaner(handlers=[SaveCleanedDataHandler(output_file='cleaned_data.csv')])
cleaned.clean_data(df.copy())
```

The above code will able us to store the cleaned data after the whole cleaning process into a csv file.

Functions of classes:

SaveCleanedDataHandler class is handler in the Chain of Responsibility, its function is to save data to the new file.
DataFrameCleaner class  function is to coordinate the cleaning process.
